## Simulaci贸n de Monte Carlo

* Generar dos n煤mero aleatorios entre 0 y 1
* Calcular x2 + y2
    * si es mayor que 1 estamos fuera del c铆rculo
    * si es menor que 1 estamos dentro del c铆rculo
* Calculamos el n煤mero total de veces que est谩n dentro del c铆rculo y lo dividimos por el n煤mero de intentos para obtener una aproximaci贸n de la probabilidad de caer dentro del c铆rculo.
* Dicha probabilidad la utilizamos para aproximar el valor de pi. (la probabilidad te贸rica es pi/4
* Repetimos el experimento un n煤mero suficiente de veces, para obtener diferentes aproximaciones de pi.
* Luego calculamos el promedio del mil experimentos para obtener un valor final.

```python
n = 100
exp = 50
pi_sum = 0
pi_value_list = []
for i in range(exp):
    value = 0
    x = np.random.uniform(0,1,n).tolist()
    y = np.random.uniform(0,1,n).tolist()
    for j in range(n):
        z = np.sqrt(x[j]*x[j]+y[j]*y[j])
        if z <=1:
            value+=1
    float_value = float(value)
    pi_value = float_value*4 / n
    pi_value_list.append(pi_value)
    pi_sum += pi_value

pi = pi_sum/exp

print(pi)
plt.plot(pi_value_list)
```
